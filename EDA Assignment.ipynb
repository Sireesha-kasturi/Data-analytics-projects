```python
Data['Total_Runs']=Data['4s']*4+Data['6s']*6
Data['Percentage']=Data['Total_Runs']/Data['Runs']*100
Data['Percentage']=Data['Percentage'].round(2)
top5=Data.sort_values('Percentage',ascending=False).head(5)
top5[['Player','Runs','4s','6s','Total_Runs','Percentage']]
```

```python
Data['Percentage_notout']=Data['Not Out']/Data['Inns']*100
Top5=Data.nlargest(5,'Percentage_notout')
Top5.loc[:,['Player','Not Out','Inns','Percentage_notout']]
```
